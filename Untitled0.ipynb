step -2 Data Cleaning Phase

```python
# Keep only necesssary column for recommendiations
movies = movies[['title', 'overview', 'genres']]
movies.head()
```

```python
# Check missing values again after column section
movies.isnull().sum()
```

```python
# Removes rows that contain missing values
movies.dropna(inplace=True)
```

```python
# reset index after removing rows
movies.reset_index(drop=True, inplace=True)
```

```python
# display cleaned dataset
movies.head()
```

Step -3 Feature Engineering Phase

```python
# icombine overview and genre into one text feature
movies['content'] = movies['overview'] + " " + movies['genres']
```

```python
# Convert All text to lowercase consistency
movies['content'] = movies['content'].str.lower()
```

```python
# Display title and combine context
movies[['title', 'content']].head()
```

step -4 Vectorization and Similarity

```python
# Import TF-IDF vectorizer to Convert text into numbers
from sklearn.feature_extraction.text import TfidfVectorizer

# Initilize and vectorzer
vectorizer = TfidfVectorizer(stop_words='english')

# Convert text content into numerical vectors
vectors = vectorizer.fit_transform(movies['content'])
```

```python
# Check shape of vectorized data
vectors.shape

# Change column size bcaz of convert (Text --->  vector --->  number) : (Text -->  Number)
```

```python
# Import cosine similarity functions
from sklearn.metrics.pairwise import cosine_similarity

# Calculate cosine similarity between vectors
similarity = cosine_similarity(vectors)
```

```python
# Check shape of simiilarity matrix
similarity.shape
```

```python
# Functions to Recommend Similar Movies
def recommend(movie_name):
  # get index of shape movie
  idx = movies[movies['title'] == movie_name].index[0]

  # get similarity scores for this movie
  scores = list(enumerate(similarity[idx]))

  # Sort movies based on similarity scores
  scores = sorted(scores, key=lambda x: x[1], reverse=True)

  # Get 5 top movie titles
  return [movies.iloc[i[0]].title for i in scores[1:6]]
```

```python
# Test Recommendiations system with a sample movie
recommend('Inception')
```
